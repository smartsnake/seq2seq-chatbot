```python
import tensorflow as tf
import tensorlayer as tl
import numpy as np
from tensorlayer.cost import cross_entropy_seq, cross_entropy_seq_with_mask
from tqdm import tqdm
from sklearn.utils import shuffle
from tensorlayer.models.seq2seq import Seq2seq
from tensorlayer.models.seq2seq_with_attention import Seq2seqLuongAttention
import os
import pandas as pd
```

```python
data_file = "data/messages.csv"
raw_data = pd.read_csv(data_file)
raw_data.tail()
```

```python
del raw_data['Emoji']
del raw_data['Attachments']
del raw_data['From Name']
del raw_data['To Name']
```

```python
data = raw_data.astype(str)
data.tail()
```

```python
len(data)
```

```python
boolean_to = data['To'].str.contains("4794049074")
boolean_from = data['From'].str.contains("4794049074")
```

```python
boolean_to.tail()
```

```python
data = data[boolean_to | boolean_from]
len(data)
```

```python
data.tail(10)
```

```python
del data["Timestamp"]
del data["Date"]
del data["Time"]
del data["From"]
```

```python
data.tail(10)
```

```python
data = data.rename({'To': 'From'}, axis=1)
data = data[1:]
data.head()
```

```python
s = (data.groupby([data['From'].ne(data['From'].shift(1)).cumsum().values, 'From'])['Message']
        .agg(' '.join).reset_index(level=1))
```

```python
s.head()
```

```python
len(s)
```

```python
data_list = []

for index in range(0, (s.shape[0])-1):
    data_list.append(s.values[index][1].replace('\n',''))
```

```python
len(data_list)
```

```python
print(data_list)
```

```python
output_file = "data/messages.txt"
with open(output_file, 'w') as f:
    for item in data_list:
        f.write("%s\n" % item, encoding="utf-8")
```
